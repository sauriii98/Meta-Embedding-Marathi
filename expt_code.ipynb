```python
def compute_similarity(data_path, model):
    def similarity(v1, v2):
        n1 = np.linalg.norm(v1)
        n2 = np.linalg.norm(v2)
        return np.dot(v1, v2) / n1 / n2

    mysim = []
    gold = []

    with open(data_path, 'rb') as fin:
        for line in fin:
            tline = line.split()
            word1 = tline[0].lower()
            word2 = tline[1].lower()

            v1 = model.get_word_vector(word1)
            v2 = model.get_word_vector(word2)
            d = similarity(v1, v2)
            mysim.append(d)
            gold.append(float(tline[2]))

    corr = stats.spearmanr(mysim, gold)
    dataset = os.path.basename(data_path)
    correlation = corr[0] * 100
    return dataset, correlation, 0
```

```python
def average_normalised_residual_error(W, Y, P):
    numerator = 0
    for w, p in zip(W, P):
        diff = Y - np.dot(w, p)
        norm = np.linalg.norm(diff)
        numerator += norm
        
    condition = (1.0 * numerator)/(r* math.sqrt(v*d))
        
    return condition
```

```python
def ensemble_SOLS(W, Y, P, theta=0.001):
    while average_normalised_residual_error(W[:], Y[:], P[:]) > theta:
        Y = Y - np.mean(Y, axis=0)
        Y = Y/np.std(Y, axis=0)
        
        Y_new = np.zeros((v, d))
        for i in range(r):
            dot_inv = np.linalg.inv(np.dot(np.transpose(W[i]), W[i]))
            dot_1 = np.dot(dot_inv, np.transpose(W[i]))
            P[i] = np.dot(dot_1, Y)
            
            Y_new = np.add(Y_new, np.dot(W[i], P[i]))
            
        Y = Y_new/r
        
    return Y
```

```python
def ensemble_SOPP(W, Y, P, theta=0.001):
    while average_normalised_residual_error(W[:], Y[:], P[:]) > theta:
        Y_new = np.zeros((v, d))
        for i in range(r):
            w_trans = np.transpose(W[i])
            S = np.dot(w_trans, Y)
            U, Ds, V_trans = np.linalg.svd(S, full_matrices=True, compute_uv=True)
            
            P[i] = np.dot(U, V_trans)
            
            Y_new = np.add(Y_new, np.dot(W[i], P[i]))
            
        Y = Y_new/r
        
    return Y
```

```python
def compute_similarity(data_path, model):
    def similarity(v1, v2):
        n1 = np.linalg.norm(v1)
        n2 = np.linalg.norm(v2)
        return np.dot(v1, v2) / n1 / n2

    mysim = []
    gold = []

    with open(data_path, 'r', encoding='utf-8') as fin:
        count = 0
        oov_count = 0
        x = 0
        for line in fin:
            tline = line.split()
            word1 = tline[0]
            word2 = tline[1]

            if (word1 in words) and (word2 in words):
                v1 = model[words.index(word1)]
                v2 = model[words.index(word2)]
                d = similarity(v1, v2)
                x += 1
            else:
                d = 0
                oov_count += 1
                
            count += 1    
            mysim.append(d)
            gold.append(float(tline[2]))
        
        print("all: ", count, "c: ", x, "oov: ", oov_count)

    corr = stats.spearmanr(mysim, gold)
    dataset = os.path.basename(data_path)
    correlation = corr[0] * 100
    return dataset, correlation, (oov_count*100.0)/count
```
